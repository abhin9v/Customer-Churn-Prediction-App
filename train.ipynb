```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
data=pd.read_csv('Churn_Modelling.csv')
```

```python
data.head()
```

```python
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
data
```

```python
label_encoder_gender=LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
data
```

```python
data['Geography'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder=OneHotEncoder()
encoded=onehotencoder.fit_transform(data[['Geography']])
encoded
```

```python
onehotencoder.get_feature_names_out(['Geography'])
```

```python
encoded_df=pd.DataFrame(encoded.toarray(),columns=onehotencoder.get_feature_names_out(['Geography']))
```

```python
encoded_df
```

```python
data=pd.concat([data,encoded_df],axis=1)
```

```python
data
```

```python
data.drop('Geography',axis=1,inplace=True)
```

```python
data
```

```python
#saving the encoder
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file )

with open('onehotencoder.pkl','wb') as file:
    pickle.dump(onehotencoder,file)
```

```python
data.head()
```

```python
x=data.drop('Exited',axis=1)
y=data['Exited']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scale=StandardScaler()
x_train=scale.fit_transform(x_train)
x_test=scale.transform(x_test)
```

```python
with open('scale.pkl','wb') as file:
    pickle.dump(scale,file)
```

```python
data
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
#ANN model
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
loss
```

```python
#compile model
model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
log_dir="logs/fit"+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
early_Stopping_callback=EarlyStopping(monitor='val_Loss',mode='min',patience=10,restore_best_weights=True)
```

```python
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,callbacks=[tensorflow_callback,early_Stopping_callback])
```

```python
model.save('model.h5')
```
